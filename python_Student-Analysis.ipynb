```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("5 - 10","05-10")
df.head()
```

```python
df1=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(df1)
```

```python
sns.heatmap(df1,annot=True)
plt.title("Students score vs Parents Education")

plt.show()
```

higher the education of the parents, the better impact on kids

```python
df2=sns.countplot(x="Gender",data=df)
plt.title("Gender Count")

df2.bar_label(df2.containers[0])
plt.show()
```

```python
df3=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(df3)
```

```python
df4=sns.heatmap(df3,annot =True)
plt.title("ParentsMarital status vs Students score")
plt.show()
```

very negligible impact on students sorce from marital status

```python
groupA = df.loc[(df["EthnicGroup"] =="group A")].count()
groupA
```

```python
groupB = df.loc[(df["EthnicGroup"] =="group B")].count()
groupC = df.loc[(df["EthnicGroup"] =="group C")].count()
groupD = df.loc[(df["EthnicGroup"] =="group D")].count()
groupE = df.loc[(df["EthnicGroup"] =="group E")].count()
g=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(g)
L=["group A","group B","group C","group D","group E"]

plt.pie(g ,labels = L, autopct ="%1.2f%%" )
plt.title("Ethnic group distribution")
plt.show()
```
